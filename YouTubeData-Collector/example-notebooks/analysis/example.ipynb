```python
from youtubecollector import FP_DATA_PROC 
import pandas as pd 
```

```python
# Specify date of creation for the processed data 
DATE = "2022-06-12"
```

```python
comments_file = FP_DATA_PROC / 'all' / f'comments-{DATE}.gzip.parquet'
video_data_file = FP_DATA_PROC / 'all' / f'video-{DATE}.gzip.parquet'

df_comments = pd.read_parquet(comments_file)
df_video = pd.read_parquet(video_data_file)
```

```python
df_comments.tail()
```

```python
df_video.tail()
```

```python
idx_most_comments = df_video.comment_counts.fillna(0).astype(int).idxmax()
video_most_comments = df_video.iloc[idx_most_comments]
video_most_comments
```

```python
df_most_comments = df_comments.query(f'video_id == "{video_most_comments.video_ids}"')
df_most_comments.tail()
```

```python
num_comments_largest_thread = df_most_comments.total_reply_count.max()
mask = df_most_comments.total_reply_count == num_comments_largest_thread
thread_id = df_most_comments.loc[mask, "thread_id"].item()
print(f"Thread id: {thread_id}", f"#Comments: {num_comments_largest_thread}", sep="\n")

# Toplevel comment and replies in largest thread
df_comments.query(f"thread_id =='{thread_id}'")  
```
