# Preprocess collected data

```python
from youtubecollector import proc_utils, data_utils
from youtubecollector.proc_utils import get_df, get_dfs, get_comment_data
```

```python
df_upload = get_df('upload', get_dfs_func=get_dfs)
df_video = get_df('video', get_dfs_func=get_dfs)
df_toplevel = get_df('toplevel', get_dfs_func=get_comment_data)
df_replies = get_df('replies', get_dfs_func=get_comment_data)

# Concat toplevel and replies 
df_comments = proc_utils.get_df_comments(df_toplevel, df_replies)  
```

```python
# Saves processed dataframes to `/data/proc/all`
names = ['video', 'comments']
dfs = [df_video, df_comments]
for df, name in zip(dfs, names):
    data_utils.save_df(df, name, folder='all')
```
